# 🔥 Firefight AI 训练笔记本
### 使用 Google Colab 免费 T4 GPU 训练 YOLO 模型

**用法：** Runtime → Run All，训练完成后模型自动保存

```
# 1. 挂载 Google Drive (保存模型)
from google.colab import drive
drive.mount('/content/drive')
```

```
# 2. 安装依赖
!pip install ultralytics -q
import torch
print(f"GPU: {torch.cuda.get_device_name(0)}")
```

```
# 3. 上传你的数据集 (把 data.yaml、images/、labels/ 打包成 zip 上传)
from google.colab import files
import zipfile, os

print("点击下方按钮上传 dataset.zip (包含 data.yaml, images/, labels/)")
uploaded = files.upload()

for fn in uploaded.keys():
    with zipfile.ZipFile(fn, 'r') as z:
        z.extractall('/content/dataset')
    print(f"已解压: {fn}")

os.chdir('/content/dataset')
!ls -la
!head -15 data.yaml
```

```
# 4. 开始训练
from ultralytics import YOLO

model = YOLO('yolov8n.pt')  # 轻量模型, T4约5分钟/epoch

results = model.train(
    data='data.yaml',
    epochs=50,
    imgsz=640,
    device=0,
    name='firefight_train',
    exist_ok=True
)

print("✅ 训练完成!")
```

```
# 5. 保存模型到 Google Drive
import shutil
best_pt = '/content/runs/detect/firefight_train/weights/best.pt'
dest = '/content/drive/MyDrive/firefight_best.pt'
if os.path.exists(best_pt):
    shutil.copy(best_pt, dest)
    print(f"✅ 模型已保存到 Google Drive: firefight_best.pt")
else:
    print("❌ 未找到 best.pt")
```
